```
import numpy as np
import plotly.graph_objects as go

x = np.linspace(0, 1)
y = np.linspace(0, 1)
xx, yy = np.meshgrid(x, y)
xy = np.array((xx.ravel(), yy.ravel()))

zz = [x*np.sin(y) for x, y in zip(xx.ravel(), yy.ravel())]
```

```
go.Figure(
    go.Scatter3d(
        x=xx.ravel(),
        y=yy.ravel(),
        z=zz,
        mode="markers"
    )
)
```
